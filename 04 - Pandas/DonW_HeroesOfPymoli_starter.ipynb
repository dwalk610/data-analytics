```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#I'm assuming this is just to count distinct players
# The line below would return a list of distinct player names
    # Player_Count = purchase_data["SN"].unique()
    # len(Player_Count)
Player_Count = purchase_data["SN"].nunique()
Player_Count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Calculate unique items using "Item ID" column
Uniq_Items = purchase_data["Item ID"].nunique()

#Calculate Average price
Avg_Price = purchase_data["Price"].mean()

#Calculate # of purchases
Purchases = purchase_data["Purchase ID"].count()
Purchases

#Calculate revenue by summing price column
Rev = purchase_data["Price"].sum()
Rev

#Create summary data frame with values calculated above
#Create a dictionary of lists
Summ_dict = {"Number of Unique Items":[Uniq_Items],
             "Average Price":[Avg_Price],
             "Number of Purchases":[Purchases],
             "Total Revenue":[Rev]
            }

#Insert dictionary of lists into data frame
Summ_dict_pd = pd.DataFrame(Summ_dict)


#Convert Average price and revenue columns to float
Summ_dict_pd["Average Price"]= Summ_dict_pd["Average Price"].astype(float).map("${:.2f}".format)
Summ_dict_pd["Total Revenue"]= Summ_dict_pd["Total Revenue"].astype(float).map("${:,.2f}".format)
Summ_dict_pd
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Get total number of unique male, female and other users
# Group purchase data by Gender
By_Gender = purchase_data.groupby(["Gender"])

# Obtain total counts by column of unique values only
    # The SN (Screenname) column will then give you the total of unique users by gender
Players_By_Gender = By_Gender.nunique()
#Players_By_Gender.reset_index(drop=True)

# # Create a new data frame which contains counts of unique values by column
By_Gender_df = pd.DataFrame(Players_By_Gender)

#Keep only the Screen Name(SN) columns
By_Gender_SN = By_Gender_df[["SN"]]

# Change SN column name to "Total Count"
By_Gender_SN = By_Gender_SN.rename(columns={"SN":"Total Count"})
By_Gender_SN

#Calculate percentage of total players by gender
Total = By_Gender_SN["Total Count"].sum()
Gender_Percent = (By_Gender_SN["Total Count"] / Total).round(2)*100
Gender_Percent

#Insert percentage column into dataframe
By_Gender_SN["Percentage of Players"] = Gender_Percent
By_Gender_SN

#Reset index back to standard 0,1,...
By_Gender_SN.reset_index()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender

* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Get total number of unique male, female and other users
# Group purchase data by Gender
Purch_Gender = purchase_data.groupby(["Gender"])

#Define purchase count
P_count = Purch_Gender["Purchase ID"].count()

#Average purchase price
P_avg = Purch_Gender["Price"].mean()
P_avg = P_avg.astype(float).map("${:.2f}".format)

#Total purchase value (revenue)
P_rev = Purch_Gender["Price"].sum()


#Calculate unique users by gender
P_players = Purch_Gender["SN"].nunique()

#Average per person
PP = P_rev/P_players
PP=PP.astype(float).map("${:.2f}".format)

#Create data frame to summarize purchase analysis results


Purch_Gen_df = pd.DataFrame({"Purchase Count":P_count,
                             "Average Purchase Price":P_avg,
                             "Total Purchase Value":P_rev.astype(float).map("${:.2f}".format),
                             "Avg Total Purchase per Person":PP})
Purch_Gen_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Define max age to set limit of age bin
max_age = purchase_data["Age"].max()
max_age

#Create age bins
bins = [0,9,14,19,24,29,34,39,max_age]

#Create names for bins
age_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#Create age summary data frame
purchase_data["Age_Summ"] = pd.cut(purchase_data["Age"], bins, labels=age_names)
purchase_data

#Group data frame by age range
purchase_age = purchase_data.groupby("Age_Summ")
purchase_age.max()

users_uniq = purchase_age["SN"].nunique()
users_uniq

#Create new dataframe
purch_age_df = pd.DataFrame({"Total Count":users_uniq})
purch_age_df

age_perc = (purch_age_df["Total Count"] / Player_Count).round(4)*100

purch_age_df["Percentage of Players"] = age_perc
purch_age_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Define max age to set limit of age bin
max_age = purchase_data["Age"].max()
max_age

#Create age bins
bins = [0,9,14,19,24,29,34,39,max_age]

#Create names for bins
age_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#Create age summary data frame
# purchase_data["Age_Summ"] = pd.cut(purchase_data["Age"], bins, labels=age_names)
# purchase_data
purchase_data

#Group data frame by age range
purchase_age = purchase_data.groupby("Age_Summ")
purchase_age.max()

#Calculate purchase count
pur_count = purchase_age["Purchase ID"].count()
pur_count.count

#Calculate average purchase price
pur_avg = purchase_age["Price"].mean()
pur_avg

#Calculate Total purchase value (revenue)
pur_tot = purchase_age["Price"].sum()
pur_tot

#Calculate average tot purchase per person
 #Calculate unique players for each age group
pur_sn = purchase_age["SN"].nunique()
pur_sn

pp = pur_tot / pur_sn
pp

#Summarize results in data frame
Purch_Age_df = pd.DataFrame({"Purchase Count":pur_count,
                             "Average Purchase Price":pur_avg,
                             "Total Purchase Value":pur_tot,
                             "Avg Total Purchase per Person":pp}
                           )
Purch_Age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results

```python
#Group data frame by screen name
purchase_mvps = purchase_data.groupby("SN")
purchase_mvps.max()

#Calculate Purchase count
pur_count = purchase_mvps["Purchase ID"].count()
pur_count

#Calculate average purchase price
pur_avg = purchase_mvps["Price"].mean()
pur_avg

#Calculate total purcahse value
pur_tot = purchase_mvps["Price"].sum()
pur_tot

#Create data frame to summarize results of top spenders
big_spender_df = pd.DataFrame({"Purchase Count":pur_count,
                               "Average Purchase Price":pur_avg,
                               "Total Purchase Value":pur_tot
                              })
big_spender_mvps = big_spender_df.sort_values("Total Purchase Value", ascending=False)
big_spender_mvps.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Retrieve Item ID, Item Name and Item Price columns and place in table
most_pop = purchase_data[["Item ID","Item Name","Price"]]
most_pop

#Group data by Item ID and Item name and Item Price
pop_group = most_pop.groupby(["Item ID","Item Name","Price"])
pop_group.count()

#Calculate purchase count
purch_count = pop_group["Item ID"].count()
purch_count


#Calculate purchase value
purch_total = pop_group["Price"].sum()
purch_total

#Create summary data frame of the results above
most_pop_df = pd.DataFrame({"Purchase Count":purch_count,
                            "Total Purchase Value":purch_total})

most_pop_df.head()
sorted_pop = most_pop_df.sort_values("Total Purchase Value", ascending=False)
sorted_pop
```
