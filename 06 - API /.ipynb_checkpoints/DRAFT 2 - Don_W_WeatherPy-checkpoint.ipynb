```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import csv
import datetime as dt

# Import API key
import api_keys
# I have removed my api key since we have to post this on GitHub. 
# I have included it in the comments section of my bootcamp spot submission
api_key = "xxxxxxxxxxxx"

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)

#Current date
today = time.strftime("%m/%d/%Y")
```

## Generate Cities List

```python
## List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#Create a variable for line break for formatting
line_break = "-----------------------------"

# Create base url
url_base = "http://api.openweathermap.org/data/2.5/weather?"

# Create full url excluding city name, which will be added in the loop
full_url = url_base + "&appid=" + api_key + "&q="
full_url

# https://samples.openweathermap.org/data/2.5/weather?q=London,uk&appid=b6907d289e10d714a6e88b30761fae22

#Create lists to hold weather data from API calls
cloud = []
country = []
humidity = []
max_temp = []
wind_speed = []
date = []
lon = []
lat = []

try:
    print("Beggining Data Retreival\n{}".format(line_break))
    
    for city in cities:
        try:
            city_data = requests.get(full_url + city).json()
            date.append(city_data["dt"])
            cloud.append(city_data["clouds"]["all"])
            country.append(city_data["sys"]["country"])
            humidity.append(city_data["main"]["humidity"])
            max_temp.append(city_data["main"]["temp_max"])
            wind_speed.append(city_data["wind"]["speed"])
            lon.append(city_data["coord"]["lon"])
            lat.append(city_data["coord"]["lat"])
            print(f"Processing Record {cities.index(city)} | {city}")
            
        except:
            print(f"City not found man sorry")
except:
    print(f"try again")

print("{}\nDate Retrieval Complete\n{}".format(line_break,line_break))
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# Create city data dictionary and convert to data frame (DF) in seperate steps
# City data lists are not the same length so trying to convert to DF right away will cause error
# Note: latitudes (lats) and longitudes (lngs) are numpy (np) arrays 
# When adding lats and lngs to dict, must convert to lists (from np array) to match other city data format 

weather_dict = {
    "Date":date,
    "Country":country,
    "City":cities,
    "Latitude":lat,
    "Longitude":lon,
    "Max Temperature":max_temp,
    "Humidity":humidity,
    "Cloudiness":cloud,
    "Wind Speed":wind_speed,
    }

# Create Data Frame (DF) from weather dictionary
# Lists are diff lenghts so use "index" for orient variable in "from_dict" 
# This will use keys as rows rather than columns when creating dict
# Transpose dataframe to set keys as columns

weather_df = pd.DataFrame.from_dict(weather_dict, orient = "index").transpose()

# Remove rows where data is missing
weather_df = weather_df.dropna()

# Write weather data to csv
weather_df.to_csv("Weather_Data.csv", index=False)
    
weather_df.head()
weather_df.count()
 
```

### Plotting the Data

#### Latitude vs. Temperature Plot

```python
# Define x and y paramters for chart

# x - axis - latitude
lat_df = weather_df["Latitude"]

# y - axis - temperature - converted from Kelvin to Fahrenheit
max_t_df = weather_df["Max Temperature"]

# Temp converted to Fahrenheit
max_t_F = (max_t_df - 273.15)*(9/5)+32
max_t_F

# Create scatter plot
plt.scatter(lat_df,max_t_F,marker = 'o',facecolors = "blue", edgecolors = "black")

# Format plot with title
plt.title(f"City Latitude Vs Max Temp ($^o$F) | {today}")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature ($^o$F)")
plt.grid(which="major",alpha = 0.3)

plt.show()

```

#### Latitude vs. Humidity Plot

```python
# Define y paramters for chart
# x axis (latitude) already defined in exercise above

# y - axis - humidity
hum_df = weather_df["Humidity"]

# Create scatter plot
plt.scatter(lat_df,hum_df,marker = 'o',facecolors = "blue", edgecolors = "black")

# Format plot with title
plt.title(f"City Latitude vs Humidity (%) | {today}")
plt.xlabel("Latitude")
plt.ylabel("Humidity(%)")
plt.grid(which="major",alpha = 0.3)

plt.show()

```

```python
# Define y paramters for chart
# x axis (latitude) already defined in exercise above

# y - axis - cloudiness
cloud_df = weather_df["Cloudiness"]

# Create scatter plot
plt.scatter(lat_df,cloud_df,marker = 'o',facecolors = "blue", edgecolors = "black")

# Format plot with title
plt.title(f"City Latitude vs Cloudiness (%) | {today}")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness(%)")
plt.grid(which="major",alpha = 0.3)

plt.show()

```

```python
# Define y paramters for chart
# x axis (latitude) already defined in exercise above

# y - axis - wind speed convert from meters/s to miles/hour
wind_df = weather_df["Wind Speed"]*2.237

# Create scatter plot
plt.scatter(lat_df,wind_df,marker = 'o',facecolors = "blue", edgecolors = "black")

# Format plot with title
plt.title(f"City Latitude vs Wind Speed (mph) | {today}")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed (mph)")
plt.grid(which="major",alpha = 0.3)

plt.show()

```
